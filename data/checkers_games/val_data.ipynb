```python
import os
import numpy as np
import torch
import pickle
#dataset = "/home/vaibhav/nanogpt/CheckersGPT/data/checkers_games"
dataset = ""
# poor man's data loader
data_dir = dataset

val_data = np.memmap(os.path.join(data_dir, 'val.bin'), dtype=np.uint8, mode='r')

file_path = '/home/vaibhav/nanogpt/CheckersGPT/data/checkers_games/meta.pkl'
with open(file_path, 'rb') as f:
    data = pickle.load(f)
print(data)

itos = data['itos']

def decode(l):
    return ''.join([itos[i] for i in l])
```
